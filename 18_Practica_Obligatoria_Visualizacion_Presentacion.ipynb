```python
df_titanic=sns.load_dataset("titanic")
df_titanic.head()
```

```python
df_titanic.drop(columns=['deck','survived','pclass','embarked'],axis=1, inplace=True)
# aquí no hace falta poner axis =1, eso es si no ponemos columns. 
df_titanic.head()
```

```python
# Parchs significa los padres o hijos con los que se viaja. siblings son hermanos y esposo, mujer, pareja etc.
# creamos columna nueva con las usma de ambas y eliminamos parch y sibs columns
df_titanic['family_members'] = df_titanic['parch'] + df_titanic['sibsp']
df_titanic.drop(columns=['sibsp','parch'],axis = 1, inplace = True)
df_titanic
```

```python
# ponemos como ciudad de embarke si no hay la moda.
df_titanic.loc[df_titanic.embark_town.isna(), "embark_town"] = df_titanic.embark_town.mode()[0]
df_titanic
```

```python
# imputamos la media de los faltantes de age según sean niño, mujer u hombre. 

es_woman = df_titanic.who == "woman"

es_child = df_titanic.who == "child"

es_man = df_titanic.who == "man"

media_woman = df_titanic.loc[es_woman,"age"].mean()

media_child = df_titanic.loc[es_child,"age"].mean()

media_man = df_titanic.loc[es_man,"age"].mean()

es_age_nulo = df_titanic.age.isna()

df_titanic.loc[es_woman & es_age_nulo, "age"] = media_woman

df_titanic.loc[es_child & es_age_nulo, "age"] = media_child

df_titanic.loc[es_man & es_age_nulo, "age"] = media_man

df_titanic
```

1- El titanic fue un accidente con una tasa de mortalidad alta.  

```python
fig,axs = plt.subplots(nrows=1,ncols=2, figsize=(10,4))
fig.suptitle("Supervivientes")
sns.countplot(x="alive", data = df_titanic, ax = axs[0], hue ="alive",legend=True,)
axs[0].set_ylabel("Número de supervivientes")
axs[0].set_title("Frecuencias Absolutas Supervivientes")
valores = df_titanic.alive.value_counts(normalize = True) * 100
sns.barplot(x = valores.index, y = valores.values, ax=axs[1],hue = valores.index,legend=True)
axs[1].set_title("Frecuencias relativas Supervivientes")
axs[1].set_ylabel("% de supervivientes");
```
