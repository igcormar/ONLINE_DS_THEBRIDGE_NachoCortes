```python
# eliminamos todas las filas con nulos. Nos cargamos bastantes filas que puede afectar al resultado, pero as lo pone en el enunciado.
cities = cities_nulos.dropna(how='any')
cities.info() # vmos que solo quedan 427 filas.
```

### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada grfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades ms pobladas, con los numeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrn, con trasparencia)
2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamao 18, color azul acero)
3. El scatter de rea de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.


```python
# las ordenamos y sacamos las 5 ms pobladas.
cities_sorted = cities.sort_values(by='population_total',ascending = False).head(5)
cities_sorted
```

```python
# creamos una lista de colores para cada barra
colores = ['blue', 'grey', 'olive', 'brown', 'bisque']
bars = plt.bar(cities_sorted['city'],cities_sorted['population_total'],color=colores)
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval)/1000, va='bottom', ha='center')

plt.title("Ciudades ms pobladas de California")
plt.xlabel("Ciudades")
plt.ylabel("Habitantes(Miles)");
```

2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamao 18, color azul acero)

```python

plt.hist(cities['area_total_km2'],50,color = 'steelblue',label = 'Hist Area')
plt.legend(loc = 'upper right', )
plt.xlabel('Superficie(km2)',fontdict={'family': 'Bradley Hand itc'},size = 18)
plt.ylabel('Nmero de ciudadanos',fontdict={'family': 'Bradley Hand itc'},size = 18);
```

3. El scatter de rea de agua de un ciudad con su poblacion.

```python


plt.scatter(cities['area_water_km2'], cities['population_total'], alpha=0.5, c = 'green');
plt.xlabel('rea de agua en la ciudad(km2)',size = 10)
plt.ylabel('Pobacin (miles habitantes)',size = 10)
plt.title("Poblacin por kilmetro cuadrado de agua")
props_flecha ={
    "arrowstyle": "->"
   
}
plt.annotate("Outliers", xy = (88, 3870000), xytext = (55,3500000), 
             arrowprops = props_flecha)
plt.annotate( "",xy = (120, 1400000),xytext = (70,3500000),arrowprops = props_flecha);
```

4. El boxplot de la altura de las ciudades.

```python
plt.boxplot(cities['elevation_m'],vert = False)
plt.title("Diagrama de caja de la Altitud de las ciudades de California")
plt.xlabel('Altitud (m)',size = 10,);
# NO S PONER LA CAJA EN ROJO.
```
